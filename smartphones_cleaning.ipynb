#Libraries

```python
import pandas as pd
```

#Importing

```python
data = pd.read_csv('smartphones.csv')
```

#Cleanup

```python
download_smartphones = False  #Set to true to download cleaned csv
```

###Multiplying resolution to get pixels

```python
cleaned = data.copy()
```

```python
cleaned.head()
```

```python
cleaned['resolution'] = cleaned['resolution_height'].astype(int) * cleaned['resolution_width'].astype(int)
cleaned.drop(['resolution_height', 'resolution_width'], axis=1, inplace=True)
```

```python
cleaned.head()
```

#Checking for null values

```python
cleaned.isnull().sum()
```

#Dropping 'fast_charging'

According the the Kaggle page, this column does not exist.

https://www.kaggle.com/datasets/abhijitdahatonde/real-world-smartphones-dataset?resource=download

```python
cleaned.drop('fast_charging', axis=1, inplace=True)
```

```python
cleaned.head()
```

```python
cleaned.isnull().sum()
```

#Dropping null values

```python
cleaned.dropna(inplace=True)
```

#Looking at result of cleaning

```python
cleaned.isnull().sum()
```

```python
print(cleaned.shape)
cleaned.head()
```

##Saving and downloading

```python
cleaned.to_csv('Cleaned smartphones.csv', index=False)
if download_smartphones:
  from google.colab import files
  files.download('Cleaned smartphones.csv')
```
