```python
import pandas as pd

data = pd.read_csv("Mobiles Dataset (2025).csv", encoding = "latin-1")

print(data.columns)
print('\n\n\n\n')
data.head()
```

##Cleanup

###Helper Functions

```python
#Price number getter function
def get_value(text):
  decimal = False
  current_num = 0
  for c in list(text):
    if c == '.':
      decimal = True
      continue

    if c.isdigit():
      current_num = current_num*10 + (ord(c) - ord('0'))
      if decimal:
        current_num = current_num/10
  return current_num


def change_to_values(data, column_id):
  data[column_id] = get_values(data[column_id])
  return data


#Function for getting values for entire column
def get_values(column):
  prices = []
  for x in column:
    prices.append(get_value(x))
  return prices

#Function for getting price values for entire column
def get_prices(column):
  return get_values(column)


#Conversion
def convert(original_list, rate):
  list = original_list.copy()
  for x in range(len(list)):
    list[x] = round(list[x]*rate, 2)
  return list


#Averages
def average_prices(original_list):
  list = original_list.copy()

  avg_prices = []
  num = len(list)

  for phone in range(len(list[0])):
    sum = 0
    for country in range(len(list)):
      sum += list[country][phone]
    avg_prices.append(round(sum/num, 2))

  return avg_prices


#Inspecting string columns for if they do not contain a specific substring
def inspect(column, substrings):
  not_found = []
  for x in column:
    if not(inspect_helper(x, substrings)):
      not_found.append(x)
  return not_found

def inspect_helper(word, substrings):
  for substring in substrings:
    if substring in word:
      return True
  return False
```

###Retreiving prices of mobiles in each country from the dataset.

```python
#Clean up
cleaned = data.copy()

#USA
USA_prices = get_prices(cleaned['Launched Price (USA)'])
print('USA = ', USA_prices)

#Pakistan
Pakistan_prices = get_prices(cleaned['Launched Price (Pakistan)'])
print('Pakistan = ', Pakistan_prices)

#India
India_prices = get_prices(cleaned['Launched Price (India)'])
print('India = ', India_prices)

#China
China_prices = get_prices(cleaned['Launched Price (China)'])
print('China = ', China_prices)

#Dubai
Dubai_prices = get_prices(cleaned['Launched Price (Dubai)'])
print('Dubai = ', Dubai_prices)
```

###Converting prices to a standard price (USD)

```python
converted_Pakistan = convert(Pakistan_prices, 0.0036)
converted_India = convert(India_prices, 0.012)
converted_China = convert(China_prices, 0.14)
converted_Dubai = convert(Dubai_prices, 0.27)
converted_USA = USA_prices

print('Pakistan = ', converted_Pakistan)
print('India = ', converted_India)
print('China = ', converted_China)
print('Dubai = ', converted_Dubai)
print('USA = ', converted_USA)
```

###Finding Average Prices

```python
prices = [converted_Pakistan, converted_India, converted_China, converted_Dubai, converted_USA]
# print(prices)
# len(prices)

avg_price = average_prices(prices)
print(avg_price)
```

###Removing prices from Cleaned dataframe and adding the Average Price

```python
del cleaned['Launched Price (USA)']
del cleaned['Launched Price (Pakistan)']
del cleaned['Launched Price (India)']
del cleaned['Launched Price (China)']
del cleaned['Launched Price (Dubai)']

cleaned['Average Price'] = avg_price

print(cleaned.columns)
print('\n\n\n')
print(cleaned.head())
```

###Separating Storage (Droping ones with no storage metric)

```python
#dropping models with no storage information
print(cleaned.shape)
values = inspect(cleaned['Model Name'], ['GB', 'TB'])
cleaned = cleaned[~cleaned['Model Name'].isin(values)]
print(cleaned.shape)
cleaned
```

```python
storage = []
for string in cleaned['Model Name']:
  i = -3
  num = 0
  digit_placement = 0
  while string[i].isdigit():
    num = num + (ord(string[i]) - ord('0')) * 10 ** digit_placement
    i = i-1
    digit_placement += 1

  if (string[-2] == 'T'):
    num = num*1024
  storage.append(num)

cleaned['Storage'] = storage
```
