#Libraries

```python
import pandas as pd
```

#Importing

```python
data = pd.read_csv('cleaned_all_phones.csv')
```

#Cleanup

```python
download_cap = False # Set to True if you want to download the cleaned data
```

###Multiplying resolution to get pixels

```python
cleaned = data.copy()
```

```python
cleaned[['resolution_height', 'resolution_width']] = cleaned['resolution'].str.split('x', expand=True)
```

```python
cleaned.head()
```

```python
cleaned['resolution'] = cleaned['resolution_height'].astype(int) * cleaned['resolution_width'].astype(int)
cleaned.drop(['resolution_height', 'resolution_width'], axis=1, inplace=True)
```

```python
cleaned.head()
```

#Checking for null values

```python
cleaned.isnull().sum()
```

##Saving and downloading

```python
cleaned.to_csv('Cleaned cleaned_all_phones.csv', index=False)
if download_cap:
  from google.colab import files
  files.download('Cleaned cleaned_all_phones.csv')
```
